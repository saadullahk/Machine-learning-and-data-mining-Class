**c)**

The transformed equation is $\frac{y_{1}^{2}}{\lambda _{1}}+\frac{y_{2}^{2}}{\lambda _{2}}=c$, which can be rewritten as:

$$\frac{y_{1}^{2}}{c\lambda _{1}}+\frac{y_{2}^{2}}{c\lambda _{2}}=1$$

The standard equation of an ellipse centered at the origin in the $(y_{1},y_{2})$ coordinate system, with semi-axes of lengths $a_{1}=\sqrt{c\lambda _{1}}$ and $a_{2}=\sqrt{c\lambda _{2}}$.

When we perform the change of variables  

$y = V^{T}(x - \mu)$,
we are doing two things to the original coordinates $x$:

- **Shift:** Subtracting $\mu$ moves the center of the coordinates to $\mu$, so the ellipse is centered at $\mu$.

- **Rotate:** Multiplying by $V^{T}$ rotates the coordinate system so that the axes line up with the ellipse direction.

The matrix $V$ contains the eigenvectors $v_{1}$ and $v_{2}$. These eigenvectors point along the principal axes of the ellipse:

 $v_{1}$: direction of the first principal axis. Its semi-axis length is $\sqrt{\lambda_{1}}$.

 $v_{2}$: direction of the second principal axis. Its semi-axis length is $\sqrt{c\,\lambda_{2}}$, where the constant $c$ sets the confidence level.

Finally, the rotation angle of the ellipse relative to the standard $x$-$y$ axes is the angle between the eigenvectors $v_{1}, v_{2}$ and the original axes.

$\textbf{K-Means:}$ is a simple clustering method that divides a dataset into 
$K$ groups based on geometric distance. It starts by selecting $K$ initial centers and then repeatedly assigns each data point to the cluster whose center is closest in Euclidean distance. After all points are assigned, the cluster centers are updated to be the mean of the points in each cluster. This process of assignment and update continues until the cluster converges. It uses Euclidean distance to define clusters, the resulting decision boundaries between clusters are linear,dividing the space into regions around each cluster. The method implicitly assumes clusters are roughly spherical, which is why a simple mean suffices to represent each cluster.

$\textbf{GMM/EM:}$ Gives points a score based on how likely they are to be under each Gaussian component. The mean and covariance of each Gaussian influence its size, shape, and direction. As a result, the high-probability areas spontaneously form ellipses. This lets GMMs represent stretched, rotated, or overlapping clusters, something K-Means cannot do.




```python
import sys
import random
import math
import numpy as np

random.seed(0)

# -------------------
# Data Set 1
# -------------------
X1 = []
for i in range(1000):
    theta = random.uniform(0, 2 * math.pi)
    radius = random.gauss(0, 0.2) + random.choice([1, 3])
    X1.append([radius * math.cos(theta), radius * math.sin(theta)])
X1 = np.array(X1)

# -------------------
# Data Set 2
# -------------------
X2 = []
for i in range(1000):
    theta = random.uniform(0, 2 * math.pi)
    radius = random.gauss(0, 0.1) + 2
    if theta < math.pi:
        X2.append([radius * math.cos(theta) - 1, radius * math.sin(theta)])
    else:
        X2.append([radius * math.cos(theta) + 1, radius * math.sin(theta)])
X2 = np.array(X2)

# -------------------
# Data Set 3
# -------------------
X3 = []
for i in range(1000):
    radius = random.gauss(0, 1)
    theta = random.uniform(0, 2 * math.pi)
    center = random.choice([[0, 1], [3, 3], [1, -3]])
    X3.append([radius * math.cos(theta) + center[0],
               radius * math.sin(theta) + center[1]])
X3 = np.array(X3)
```

```python
import matplotlib.pyplot as plt

# Plot the three datasets side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].scatter(X1[:, 0], X1[:, 1], s=10, c="blue")
axes[0].set_title("Data Set 1")
axes[0].set_xlabel("Dimension 1")
axes[0].set_ylabel("Dimension 2")

axes[1].scatter(X2[:, 0], X2[:, 1], s=10, c="green")
axes[1].set_title("Data Set 2")
axes[1].set_xlabel("Dimension 1")
axes[1].set_ylabel("Dimension 2")

axes[2].scatter(X3[:, 0], X3[:, 1], s=10, c="red")
axes[2].set_title("Data Set 3")
axes[2].set_xlabel("Dimension 1")
axes[2].set_ylabel("Dimension 2")

plt.tight_layout()
plt.show()
```

```python
## assume 'labels' is a list/array of group labels
## and X is the dataset being clustered
import matplotlib.pyplot as plt
import numpy as np

def plot_clusters(X, labels, title="Clustering Result"):
    """
    X: np.array of shape (n_samples, 2)
    labels: cluster labels (list or np.array)
    """
    unique_labels = np.unique(labels)
    colors = ['r', 'b', 'y', 'g', 'c', 'm', 'k', 'orange']

    for i, label in enumerate(unique_labels):
        points = X[labels == label]
        plt.scatter(points[:, 0], points[:, 1],
                    c=colors[i % len(colors)],
                    label=f"Cluster {label}", s=10)

    plt.xlabel("Dimension 1")
    plt.ylabel("Dimension 2")
    plt.title(title)
    plt.legend()
    plt.show()
```
